```python
#Sara Sofia Gonzalez Mora
#Taller Espacios Vectoriales 

#Ejercicio 1
#Complex  internal product for column vectors

import numpy as np

a = np.array([1+2j, 3-4j], dtype=complex)
b = np.array([2-1j, -1+3j], dtype=complex)

#Sumar los vectores
a + b
#Inverso
-a
#Escalar
c = 2+1j
c * a

#Ahora el producto interno de los dos vectores
PRODUCTO_INTERNO = np.vdot(a, b) #Funcion que conjuga los vectores
PRODUCTO_INTERNO

#Ejercicio 2
#Complex internal product for square complex matrices

A = np.array([[1+1j, 2-2j],
              [3+3j, 4-j]], dtype=complex)

B = np.array([[1-2j, 3+j],
              [4-4j, 2+2j]], dtype=complex)

producto_interno = np.vdot(A, B)
producto_interno

#Excercises on Eigenvalues and Eigenvectors 
#Ejercicio 1
#Compute Eigenvalues and Eigenvectors of a real matrix

A = np.array([[2, 1],
             [2, 3]])

#Valores y vectores propios

valor_A, vector_A = np.linalg.eig(A)
valor_A
vecotr_A

#Ejercicio 2
#Eigenvalues and Eigenvectors of a complex matrix

B = ([[1+2j 2+3j],
      [4+5j 6+7j]], dtype=complex)

#Valores y vectores propios
valor_B, vector_B = np.linalg.eig(B)

valor_B
vector_B

#Ejercicio 3
#Symmetric Matrix Eigendecomposition

C = np.array([[5, 4],
              [4, 5]])

#Valores y vectores propios
valor_C, vector_C = np.linalg.eig(C)

valor_C
vector_C

#Ejercicio 4
# Plotting Matrix Transformations

T = np.array([[2,1],
              [1,3]])

#Plot unit circle
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 200)
x = np.cos(theta)
y = np.sin(theta)

circle = np.vstack((x, y))

#Transformar
transformed = T @ circle

#Valores y vectores propios
valor_T, vector_T = np.linalg.eig(T)

#Show the graphic
plt.figure(figsize=(6,6))

plt.plot(circle[0], circle[1], label="Círculo unidad")
plt.plot(transformed[0], transformed[1], label="Transformado")

for i in range(2):
    v = vector_T[:, i]
    plt.arrow(0,0, v[0], v[1], color='red', width=0.02)

plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()
plt.axis('equal')
plt.grid()
plt.show()

#Ejercicio 5
#Diagonal Matrix Eigendecomposition
D = np.array([[7, 0],
              [0, -3]])

valor_D, vector_D = np.linalg.eig(D)

valor_D
vector_D

```
